# 01 - Data Exploration

Quick look at EmpatheticDialogues: dataset stats, emotion distribution, and cluster assignments.

```python
import sys
sys.path.insert(0, "..")

from collections import Counter
from src.data.load import load_splits
from src.data.cluster import build_emotion_clusters, save_clusters
from src.data.format import format_for_sft
```

## Load dataset

```python
splits = load_splits()
for name, convs in splits.items():
    print(f"{name}: {len(convs)} conversations")
```

## Emotion distribution

```python
emotions = [c["emotion"] for c in splits["train"]]
counts = Counter(emotions)
for emotion, count in counts.most_common():
    print(f"{emotion:20s} {count:5d}")
```

## Cluster assignments

```python
cluster_result = build_emotion_clusters()
for name, info in cluster_result["clusters"].items():
    print(f"\n{name} (id={info['id']}):")
    print(f"  centroid: V={info['centroid']['valence']:.3f}, A={info['centroid']['arousal']:.3f}, D={info['centroid']['dominance']:.3f}")
    print(f"  emotions: {info['emotions']}")
```

## Conversations per cluster

```python
emotion_to_cluster = cluster_result["emotion_to_cluster"]
cluster_names = cluster_result["cluster_names"]

cluster_counts = Counter(emotion_to_cluster[c["emotion"]] for c in splits["train"])
for cid, count in sorted(cluster_counts.items()):
    print(f"Cluster {cid} ({cluster_names[cid]}): {count} conversations")
```

## Sample formatted training example

```python
examples = format_for_sft(splits["train"][:10], emotion_to_cluster)
ex = examples[0]
print(f"emotion: {ex['emotion']}")
print(f"cluster_id: {ex['cluster_id']}")
print(f"conv_id: {ex['conv_id']}")
print(f"\n--- context ---")
print(ex["context"])
print(f"\n--- response ---")
print(ex["response"])
print(f"\n--- full text ---")
print(ex["text"])
```
